```python
from pandas import Series, DataFrame
```

```python
s = Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
```

```python
df = DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
```

```python
df.plot()
```
